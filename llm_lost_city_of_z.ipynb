**Creating a Test Dataset**
- One way to evaluate it is by building a test dataset—a small, controlled set of images where we already know the ground truth. For example:
    - 10 images with known historical sites
    - 10 images without historical sites

- This lets us check if the model can correctly identify which images have sites and which don't.

**Understanding Accuracy**

- Accuracy is simply the percentage of times the model gets it right overall.
- If out of 20 images, it labels 18 correctly, accuracy is: 18/20=90%

⚠️ But accuracy can be misleading.
- Imagine our test dataset is very imbalanced. Say 95 images have no historical site. Only 5 images do have a site. A really dumb model could just always predict “no site” and still get 95% accuracy. But it would miss every actual archaeological site! That’s why accuracy alone isn’t enough for many real-world problems, especially when one class is rare.

**Introducing Precision and Recall**
- To better understand how well our model finds the rare, important sites, we use precision and recall.

**Precision**
- Precision answers:

    - When the model says "this is a site," how often is it correct?

- High precision means few false alarms. The model is careful about saying something is a site.

**Recall**
- Recall answers:

    - Of all the actual sites, how many did the model find?

- High recall means few misses. The model is good at finding everything, even if it makes more mistakes.

**Tradeoffs Between Precision and Recall**
- Often, there’s a tradeoff.
    - A high precision model might be very strict, only flagging sites when it's very sure. But it might miss some real sites → low recall.
    - A high recall model might catch almost every site, but also produce many false positives → low precision.

**Which Do We Want?**
- Now think about our real-world scenario. Sending scientists to verify a site is very expensive. We don't want to waste time and money checking lots of false positives. So in this context, we would favor a high precision, even if it means lower recall.

- We’d rather the model only flags sites it's really confident about—even if it misses some—so that when scientists go to verify, they have a much higher chance of finding something real.

**Summary**
- Accuracy measures overall correctness—but can be misleading if classes are imbalanced.

- Precision = How reliable are positive predictions?

- Recall = How good is the model at finding all positives?

- In our archaeological search:

    - High precision = fewer false alarms, better use of expert time.

    - Lower recall is acceptable if it saves costly expeditions.

```python
from pydantic import BaseModel
from tqdm import tqdm

# We want to restrict output to a json format with two fields: contain_site yes/no and explanation
class ExamineResult(BaseModel):
  contain_site: bool
  explanation: str

def examine_site(image_url):
  im = Image.open(requests.get(image_url, stream=True).raw)
  prompt = """
  You are an expert for discovering archeology sites.
  Tell me if there is a archeological site in the given satellite image.
  Tell me how did you reach the conclusion.
  """

  result = client.models.generate_content(
          model="gemini-1.5-pro",
          contents = [prompt, im],
          config = {
              "response_mime_type": "application/json",
              "response_schema": ExamineResult,
              },
  )
  r = result.parsed
  return r

def dataset_gen():
  known_sites_fn = ['amapa.jpg',
                    'amar8.jpg',
                    'amas1.jpg',
                    'amaq53.jpg',
                    'amazd.jpg',
                    'acrr20.jpg',
                    'amaq25.jpg',
                    'amats.jpg',
                    'rpars.jpg',
                    'acrq14.jpg']

  random_pics_fn = ['arcs1.jpg',
                    'arcs2.jpg',
                    'arcs3.jpg',
                    'arcs4.jpg',
                    'arcs5.jpg',
                    'arcs6.jpg',
                    'arcs7.jpg',
                    'arcs8.jpg',
                    'arcs9.jpg',
                    'arcs10.jpg']

  known_sites_base = "https://storage.googleapis.com/arcknow-test-public/known_sites_2/"
  random_pics_base = "https://storage.googleapis.com/arcknow-test-public/random_pics/"

  dataset = []
  for p in known_sites_fn:
    dataset.append((known_sites_base+p, True))
  for p in random_pics_fn:
    dataset.append((random_pics_base+p, False))
  return dataset

def calc_metrics(exam_func):
  dataset = dataset_gen()
  TP = 0 # TP: Predicted site & is site
  TN = 0 # FP: Predicted site but no site
  FP = 0 # TN: Predicted no site & is no site
  FN = 0 # FN: Predicted no site but is site
  for d in tqdm(dataset):
    r = exam_func(d[0])
    TP += (r.contain_site and d[1])
    TN += (not r.contain_site) and (not d[1])
    FP += r.contain_site and (not d[1])
    FN += (not r.contain_site) and d[1]
  precision = TP/(TP + FP) # Calculate precision from TP and FP
  recall = TP/(TP + FN) # Calculate recall from TP and FN
  return {"precision": precision, "recall": recall, "TP": TP, "TN": TN, "FP": FP, "FN": FN}

m = calc_metrics(examine_site)
print("\n")
print(m)
```

  So how do we teach it?
  Enter few-shot learning.

  Few-shot learning is a way to give the model a small number of examples right in the prompt.
  We don’t retrain the entire model from scratch—instead, we guide it by showing patterns from known samples.

  Here’s a simple example from text summarization:
  Suppose we want the model to generate summaries in a specific style.
  In our prompt, we can include sample summaries.
  The model sees these few examples and learns to mimic the style and structure in its new answer.

  This approach works surprisingly well because LLMs are very good at pattern matching.
  Even a few clear examples can dramatically improve output quality.

**Few-shot learning for text summarization**

Here’s a simple example from text summarization:
Suppose we want the model to generate summaries in a specific style.
In our prompt, we can include:

- Example 1: Article ➜ Summary
- Example 2: Article ➜ Summary
- Then, “Now summarize this new article.”

The model sees these few examples and learns to mimic the style and structure in its new answer.

**Few-shot learning using multimodal samples**

Now let’s connect this to our lab problem.

Instead of text-only prompts, we’ll be using multimodal samples.
Our goal: detecting archaeological sites in satellite imagery.

In this lab, we’ll experiment with few-shot learning by providing the LLM with example satellite images:
- Images that contain known archaeological sites
- Images that don’t contain sites

By seeing these labeled examples in the prompt, the model can learn the distinguishing patterns:
- Artificially shaped depressions
- Geometric clearing patterns
- Symmetry not found in nature

Then, when we give it a new, unseen satellite image, the LLM will try to apply those learned patterns to decide if it contains similar archaeological features.

**Summary**
- Our baseline prompt failed because it lacked local knowledge.
- Few-shot learning lets us inject this knowledge by showing examples.
- It’s an efficient way to customize a powerful pre-trained model for our specific archaeological detection task.

```python
def examine_fewshot(image_url):
  im = Image.open(requests.get(image_url, stream=True).raw)
  sample0_url = "https://storage.googleapis.com/arcknow-test-public/known_sites/abuna_oval.jpg"
  sample1_url = "https://storage.googleapis.com/arcknow-test-public/known_sites/acrs_22.jpg"
  sample2_url = "https://storage.googleapis.com/arcknow-test-public/random_pics/arcs6.jpg"
  sample0 = Image.open(requests.get(sample0_url, stream=True).raw)
  sample1 = Image.open(requests.get(sample1_url, stream=True).raw)
  sample2 = Image.open(requests.get(sample2_url, stream=True).raw)

  prompt0 = """
  The above satellite images contain archeological site evidence.
  Close to the center of the images, there are unusal square shapes.
  The land is depressed on the edge of the square outline.
  The following image does not contain archelogical site evidence.
  Refer to this image to minimize false positive.
  """

  prompt1 = """
  Examine the following picture closely and tell me
  if you observed similar archeological evidence as the above given images.
  In this prediction, you want to minimize false positive.
  """

  result = client.models.generate_content(
          model="gemini-1.5-pro",
          contents = [sample0, sample1, prompt0, sample2, prompt1, im],
          config = {
              "response_mime_type": "application/json",
              "response_schema": ExamineResult,
              },
  )
  r = result.parsed
  return r

m = calc_metrics(examine_fewshot)
print("\n")
print(m)
```
